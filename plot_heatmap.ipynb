```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

mask = np.zeros((4,4))
mask[0,:3] = 1

# 0.849
map = np.array([[0, 0, 0, 0.870],
       [0.862, 0.870, 0.870, 0.875],
       [0.867, 0.858, 0.874, 0.876],
       [0.865, 0.867, 0.862, 0.865]]) *100

# 0.958
map50 = np.array([[0, 0, 0, 0.961],
       [0.955, 0.964, 0.963, 0.966],
       [0.962, 0.951, 0.964, 0.967],
       [0.957, 0.960, 0.957, 0.959]]) *100

# 0.894
mar = np.array([[0, 0, 0, 0.902],
       [0.900, 0.906, 0.906, 0.906],
       [0.901, 0.890, 0.911, 0.909],
       [0.901, 0.896, 0.898, 0.901]]) *100

sns.set(font_scale = 1.5)
cmap = mpl.cm.get_cmap("RdBu_r").copy()
cmap.set_bad("white")

plt.figure(figsize=(21,5))
plt.tight_layout()

maps = [map, map50, mar]
titles = ['mAP','mAP@.5','mAR']
ticks = [[81,84.9,88],[94.5,95.8,97.5],[89,89.4,91.5]]
num_heats = 3
for idx in range(num_heats):
    plt.subplot(1,num_heats, idx+1)
    res = sns.heatmap(maps[idx],mask=mask,
           annot=True, fmt='.1f', annot_kws={"fontsize":15},
           cmap=cmap,square=True, linewidth=2,
           yticklabels=['L=1','L=2','L=3','L=4'],
           xticklabels=['beta=1e-3','beta=1e-2','beta=1e-1','beta=1'],
           cbar_kws={"ticks":ticks[idx]},
           vmin=ticks[idx][0], vmax=ticks[idx][2], center=ticks[idx][1])
    plt.title(titles[idx])
    res.set_yticklabels(res.get_yticklabels(), rotation=0)
    res.set_xticklabels(res.get_xticklabels(), rotation=30)
    #res.axhline(y=-0.01, color='k',linewidth=1)
    #res.axhline(y=3.98, color='k',linewidth=1)
    #res.axvline(x=-0.01, color='k',linewidth=1)
    #res.axvline(x=3.99, color='k',linewidth=1)
```
